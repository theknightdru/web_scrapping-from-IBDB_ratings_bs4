```python
#here i am requesting the contents of a single page
from requests import get
url = 'http://www.imdb.com/search/title?release_date=2020&sort=num_votes,desc&page=1'
response = get(url)
print(response.text[:500])
```

```python
#we will import the bs library for parsing the html code
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#getting all the div containers that have a class atribute
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(movie_containers))
print(len(movie_containers))
```

### Now we’ll select  each item of interest:

* The name of the movie.
* The year of release.
* The IMDB rating.
* The Metascore.
* The number of votes.

```python
#declaring the first movie container

first_movie = movie_containers[0]
```

```python
#getting the first movie name
first_name = first_movie.h3.a.text
first_name
```

```python
#getting year of release 
first_year = first_movie.h3.find('span', class_ = 'lister-item-year text-muted unbold')
first_year.text
```

```python
#getting rating of the movie
first_imdb = float(first_movie.strong.text)
first_imdb
```

```python
#getting the metascore
first_mscore = first_movie.find('span', class_ = 'metascore mixed')
first_mscore = int(first_mscore.text)
print(first_mscore)
```

```python
#getting the votes 
first_votes = first_movie.find('span', attrs = {'name':'nv'})
first_votes = int(first_votes['data-value'])
first_votes
```

```python
#as we have the codes of html for each attribute lets get the data of the entire webpage
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
# Extract data from individual movie container
for container in movie_containers:
# If the movie has Metascore, then extract:
    if container.find('div', class_ = 'ratings-metascore') is not None:
        #name
        name = container.h3.a.text
        names.append(name)
        #year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
        #rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        #metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
        #votes
        vote = container.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
```
